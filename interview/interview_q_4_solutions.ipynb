### 1. Pythondaki farklı operator tipleri nelerdir?

**1. Aritmetik Operatörler:** Sayısal değerler üzerinde matematiksel işlemler yapmak için kullanılır.

Operatör	Açıklama	Örnek

* "+"   	Toplama 	5 + 2 = 7

* "-"	Çıkarma	5 - 2 = 3

* "*"	Çarpma	5 * 2 = 10

* "/"	Bölme	5 / 2 = 2.5

* "//"	Tam Sayı Bölmesi	5 // 2 = 2

* "%"	Mod Alma (Kalan)	5 % 2 = 1

* "**"	Üs Alma	5 ** 2 = 25


**2. Karşılaştırma Operatörleri:** Değerleri karşılaştırmak ve True (Doğru) veya False (Yanlış) değerleri döndürmek için kullanılır.

Operatör	Açıklama	Örnek

* "=="	Eşit mi	5 == 5 (True)

* "!="	Eşit değil mi	5 != 5 (False)

* ">"	Büyük mü	5 > 2 (True)

* "<"	Küçük mü	5 < 2 (False)

* ">="	Büyük veya Eşit mi	5 >= 5 (True)

* "<="	Küçük veya Eşit mi	5 <= 2 (False)


**3. Atama Operatörleri:** Değişkenlere değer atamak için kullanılır.

Operatör	Açıklama	Örnek

* "="	Atama	x = 5

* "+="	Topla ve Ata	x += 5 (x = x + 5)

* "-="	Çıkar ve Ata	x -= 5 (x = x - 5)

* "*="	Çarp ve Ata	x *= 5 (x = x * 5)

* "/="	Böl ve Ata	x /= 5 (x = x / 5)

* "//="	Tam Sayı Böl ve Ata	x //= 5 (x = x // 5)

* "%="	Modunu Al ve Ata	x %= 5 (x = x % 5)

* "**="	Üssünü Al ve Ata	x **= 5 (x = x ** 5)


**4. Mantıksal Operatörler:** Mantıksal ifadeleri birleştirmek için kullanılır.

Operatör	Açıklama	Örnek

* "and"	Ve	(5 > 2) and (10 < 20) (True)
* "or"	Veya	(5 > 2) or (10 > 20) (True)
* "not"	Değil	not(5 > 2) (False)

### 2. Pythonda map() fonksiyonu nedir?

map() fonksiyonu, Python'da bir iterable (liste, tuple, dize vb.) üzerinde bir fonksiyonu çalıştırmak için kullanılan bir fonksiyonel programlama aracıdır. Her bir elemanı tek tek alır, verilen fonksiyonu uygular ve sonuçları içeren yeni bir iterable döndürür.

```python
def kare_al(x):
  return x ** 2

sayilar = [1, 2, 3, 4, 5]

kareler = map(kare_al, sayilar)
print(list(kareler)) 
```

```python
#Çıktısı map nesnesi olduğu için yazdırırken listeye dönüştürülür.
def kare_al(x):
  return x ** 2

sayilar = (1, 2, 3, 4, 5)

kareler = map(kare_al, sayilar)
print(kareler)
```

### 3. Pythonda özyineleme (recursion) nedir?

Recursion bir fonksiyonun kendisini içerisinde çağırmasıdır.

```python
def faktoriyel(n):
  """
  Verilen bir sayının faktöriyelini hesaplayan rekürsif fonksiyon.
  """
  if n == 0: # Temel durum
    return 1
  else: # Rekürsif adım
    return n * faktoriyel(n-1)

print(faktoriyel(5))
```

### 8. Mutable ve immutable veri tipleri arasındaki fark nedir?

Mutable veri tipleri, içerdikleri verileri değiştirebilirken, immutable veri tipleri değiştirilemez.

Mutable veri tipleri üzerinde yapılan değişiklikler orijinal nesneyi etkilerken, immutable veri tipleri üzerinde yapılan değişiklikler yeni bir nesne oluşturur.

immutable veriler: 
Sayılar (int, float, complex),
Stringler (str),
Tuple'lar (tuple)

mutable veriler:
Listeler (list),
Sözlükler (dict),
Kümeler (set)


```python
my_list = [1, 2, 3]
my_list[0] = 4  # Liste değiştirildi
print(my_list) 
```

```python
my_string = "Merhaba"
my_string[0] = "S"  # Hata verir: Dizeler değiştirilemez
print(my_string)
```

```python
my_string = "Hello"
new_string = my_string.replace("H", "J")  # Yeni bir string oluşturuluyor
print(my_string)  # Çıktı: Hello (değişmedi)
print(new_string) 
```

### 9. Pythonda / ve // arasındaki fark nedir?

/ operatörü, iki sayıyı böler ve sonucu her zaman float (ondalık) olarak döndürür.

// operatörü, iki sayıyı böler ve sonucu tam sayı (integer) olarak döndürür. Yani, bölme işleminin sonucunun tam kısmını alır.

```python
print(-5 / 2)   
print(-5 // 2)
```

### 10. PIP nedir?

PIP, Python Paket Yöneticisi'nin (Python Installer Package) kısaltmasıdır. Python programlama dilinde kullanılan paketlerin ve modüllerin indirilmesi, kurulması ve yönetilmesi için kullanılan bir araçtır.

PIP komutları:

**Paket Yükleme:** pip install <paket_adı>

**Paket Güncelleme:** pip install --upgrade <paket_adı>

**Paket Kaldırma:** pip uninstall <paket_adı>
